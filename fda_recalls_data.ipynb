```python
# Import libraries

import pandas as pd
from pandas.io.json import json_normalize
import json
import numpy as np
```

## Convert JSON to CSV

```python
# Load json. Convert to pandas dataframe

with open('device-recall-0001-of-0001.json', 'r') as handle1:
    parsed1 = json.load(handle1)
df1 = json_normalize(parsed1["results"], record_prefix = True)

with open('device-enforcement-0001-of-0001.json', 'r') as handle2:
    parsed2 = json.load(handle2)
df2 = json_normalize(parsed2["results"], record_prefix = True)
```

```python
# Create csv files

df1.to_csv("device-recall.csv")
df2.to_csv("device-enforcement.csv")
```

```python
# Explore files

print("Device recalls: \n", df1.columns, "\n")
print("Device enforcement: \n", df2.columns)
```

```python
# Size of dataframes

print(df1.shape)
print(df2.shape)
```

## Explore Enforcement dataframe

```python
# Explore differences between multiple rows with same recall numbers in Enforcement dataframe

df2temp = df2[df2["recall_number"] == "Z-1006-2014"]
df2temp
```

```python
# Print actual differences

for col in df2temp.columns:
    if df2temp[col][3884] != df2temp[col][27115]:
        print(col, df2temp[col][3884], "*********", df2temp[col][27115], "*********")
```

```python
# There are different rows for status - ongoing, terminated, and completed
```

```python
# Explore differences between multiple rows with same event IDs in Enforcement dataframe

df2temp = df2[df2["event_id"] == "66500"]
df2temp
```

```python
# Code_info, product_description, recall_number, status, termination_date are all different
```

```python
# Create new enforcement dataframe containing only rows where termination_date is not null

df2_term = df2[~df2["termination_date"].isnull()]
print("Original dataframe: ", df2.shape)
print("New dataframe: ", df2_term.shape)
```

```python
# Check if recall numbers are all unique in new enforcement dataframe

df2_term["recall_number"].value_counts()
```

```python
# Check some examples

df2_term[df2_term["recall_number"] == "Z-0827-2014"]
```

```python
# Still some differences. Explore further

for col in df2_term.columns:
    if df2_term[col][7211] != df2_term[col][18249]:
        print(col, df2_term[col][7211], "*********", df2_term[col][18249], "*********")
```

```python
# Remove duplicates, keep only last row

df2_term_unique = df2_term.drop_duplicates("recall_number", keep = "last")
print("Original dataframe: ", df2_term.shape)
print("New dataframe: ", df2_term_unique.shape)
```

```python
# Check if recall numbers are all unique in new enforcement dataframe

df2_term_unique["recall_number"].value_counts()
```

## Merge Enforcement and Recall dataframes

```python
df3 = pd.merge(df2_term_unique, df1_unique, how = "left", right_on = "product_res_number", left_on = "recall_number")
df3.shape
```
